```python
import numpy as np
import pandas as pd
import math
import time
from shapely.geometry import Point
from shapely.geometry.polygon import Polygon
```

```python
ohrc_df = pd.read_csv(r"csv_files/ohrc.csv")
```

```python
tmc_df = pd.read_csv("csv_files/tmc_oth.csv")
```

```python
tmc_rows_ohrc_cols = pd.DataFrame(index=range(855),columns=range(90))
```

```python

count_intersects =0
count_contains = 0
points =[[] for k in range(90)]
truthvalue = [False,False,False,False]

for i in range(90):
    points[i]=[Point(ohrc_df.ul_la[i], ohrc_df.ul_lo[i]), Point(ohrc_df.ur_la[i], ohrc_df.ur_lo[i]), Point(ohrc_df.ll_la[i], ohrc_df.ll_lo[i]), Point(ohrc_df.lr_la[i], ohrc_df.lr_lo[i])]

polygons = []
for i in range(855):
    polygons.append(Polygon([(tmc_df['ul_la'][i], tmc_df['ul_lo'][i]), (tmc_df['ur_la'][i], tmc_df['ur_lo'][i]), (tmc_df['lr_la'][i], tmc_df['lr_lo'][i]), (tmc_df['ll_la'][i], tmc_df['ll_lo'][i])]))
for k in range(855):
    for i in range(90):
        for j in range(4):
            truthvalue[j] = polygons[k].contains(points[i][j])
        tmc_rows_ohrc_cols.iloc[k,i]="n"
        if(truthvalue[0]==True or truthvalue[1]==True or truthvalue[2]==True or truthvalue[3]==True):
            tmc_rows_ohrc_cols.iloc[k,i]="intersects"
            count_intersects+=1
        if(truthvalue[0]==True and truthvalue[1]==True and truthvalue[2]==True and truthvalue[3]==True):
            tmc_rows_ohrc_cols.iloc[k,i]="contains"
            count_contains+=1
            count_intersects-=1
```
